```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import string
```

```python
# Descargar recursos de nltk
nltk.download('punkt')
nltk.download('stopwords')

# Dataset de ejemplo (puedes reemplazarlo con un dataset real como IMDB o SMS Spam)
data = {
    'text': [
        "I love this movie, it's amazing!",
        "This film is terrible, waste of time.",
        "Best movie ever, highly recommend!",
        "Awful acting and bad story.",
        "Great film, I really enjoyed it!",
        "Not good, very disappointing.",
    ],
    'label': [1, 0, 1, 0, 1, 0]  # 1 = positivo, 0 = negativo
}
df = pd.DataFrame(data)

# Preprocesamiento: limpieza básica
def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = word_tokenize(text)
    words = [word for word in words if word not in stopwords.words('english')]
    return ' '.join(words)

df['clean_text'] = df['text'].apply(preprocess_text)
```

```python
# Representación de texto con Bag of Words y TF-IDF
vectorizer_bow = CountVectorizer()
vectorizer_tfidf = TfidfVectorizer()

X_bow = vectorizer_bow.fit_transform(df['clean_text'])
X_tfidf = vectorizer_tfidf.fit_transform(df['clean_text'])

y = df['label']
X_train_bow, X_test_bow, y_train, y_test = train_test_split(X_bow, y, test_size=0.2, random_state=42)
X_train_tfidf, X_test_tfidf, _, _ = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
```

```python
# Modelos de clasificación
models = {
    "Multinomial Naïve Bayes": MultinomialNB(),
    "Bernoulli Naïve Bayes": BernoulliNB(),
    "Logistic Regression": LogisticRegression(),
    "Support Vector Machine": SVC(),
    "Random Forest": RandomForestClassifier()
}

# Entrenar y evaluar modelos
def train_and_evaluate(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"\n{model.__class__.__name__}:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
    plt.show()
# Pruebas con BoW y TF-IDF
for name, model in models.items():
    print(f"=== {name} (BoW) ===")
    train_and_evaluate(model, X_train_bow, X_test_bow, y_train, y_test)
    print(f"=== {name} (TF-IDF) ===")
    train_and_evaluate(model, X_train_tfidf, X_test_tfidf, y_train, y_test)
```
